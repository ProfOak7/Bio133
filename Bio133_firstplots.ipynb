```python
# importing the required module 
import matplotlib.pyplot as plt 
  
# x axis values 
x = [0,0.5,1,2,3, 4, 5,6,7,8] 

# corresponding y axis values 
y = [80, 80, 122,130,135,125,100,92,83,81] 
  
# plotting the points  
plt.plot(x, y) 
  
# naming the x axis 
plt.xlabel('time (min)') 
# naming the y axis 
plt.ylabel('heart rate (bpm)') 
  
# giving a title to my graph 
plt.title('Heart rate during exercise') 
  
# function to show the plot 
plt.show() 


```

```python
#Now let's try importing an excel document using pandas (and openpyxl) - note: df is short for dataframe
import pandas as pd

#Notice in the following commend I also set the first column (0) to be the index column. This will be helpful when trying to make heat maps later... 

df = pd.read_excel (r'/Users/jonathanokerblom/Lungsep_stu.xlsx', index_col=0) #Be sure to specify the actual location of the file - place "r" before the path string to address special character, such as '\'.
print (df)
```

```python
#Now let's make a heatmap of the data using seaborn:
import seaborn as sb

sb.heatmap(df)
plt.show()
```

```python
#We can specify the figure size using matplotlib.plyty (plt)
fig1 = plt.subplots(figsize=(10, 10)) 
sb.heatmap(df)
plt.show()
```

```python
#We can change the color and add numerical notation to each cell as shown below:
fig1 = plt.subplots(figsize=(10, 15)) 
sb.heatmap(df, cmap = 'Blues', annot = True)
plt.show()
```

```python
#We can make the color scale more robust, add a line in between each cell, and change the size of the color indicator bar... 
fig1 = plt.subplots(figsize=(10, 15)) 
sb.heatmap(df, cmap = "YlGnBu", annot = True, robust = True, linewidth=0.1, cbar_kws={"shrink": .5})
plt.show()
```
